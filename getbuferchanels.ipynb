```python
# Buffer replaced their old REST API with a GraphQL API at api.buffer.com.
# Auth is a personal Bearer token — no OAuth needed for your own account.
# Get your key at: https://publish.buffer.com/settings/api
#
# The config needs channel IDs, not profile names. Run this once to find them:
#   python3 -c "
import requests, yaml
cfg = yaml.safe_load(open('config.yaml'))
key = cfg['posting']['buffer']['api_key']
r = requests.post('https://api.buffer.com',
    headers={'Authorization': f'Bearer {key}', 'Content-Type': 'application/json'},
    json={'query': '{ account { channels { id name service } } }'},
    timeout=15)
for ch in r.json()['data']['account']['channels']:
    print(ch['id'], ch['service'], ch['name'])
    
```
